## Report Page

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

def load_lighthouse_json(file_path):
    """Loads Lighthouse JSON and extracts network request data."""
    with open(file_path, "r", encoding="utf-8") as file:
        data = json.load(file)

    # Extract network requests
    network_requests = data["audits"]["network-requests"]["details"]["items"]
    
    # Convert data to DataFrame
    df = pd.DataFrame(network_requests)

    # Convert transferSize from bytes to KB
    df["transferSizeKB"] = df["transferSize"] / 1024

    return df

def filter_localhost_scripts(df):
    """Filters only script requests from localhost:3000 and localhost:8080 and sorts by size."""
    filtered_df = df[
        (df["resourceType"] == "Script") & 
        (df["url"].str.startswith(("http://localhost:3000", "http://localhost:8080")))
    ].copy()

    # Sort by transfer size descending
    filtered_df = filtered_df.sort_values(by="transferSizeKB", ascending=False)

    return filtered_df

def plot_top_localhost_scripts(script_df):
    """Bar chart for top script files from localhost by transfer size."""
    plt.figure(figsize=(12, 6))
    plt.barh(script_df["url"], script_df["transferSizeKB"], color="blue")
    plt.xlabel("Transfer Size (KB)")
    plt.ylabel("Script URL")
    plt.title("Top Script Files from localhost:3000 and localhost:8080 by Data Usage")
    plt.gca().invert_yaxis()  # Invert y-axis to have the largest on top
    plt.grid(axis="x", linestyle="--", alpha=0.7)
    plt.show()

def main():
    # Provide the path to your Lighthouse JSON report
    file_path = "lighthouse_report_page.json"

    # Load data
    df = load_lighthouse_json(file_path)

    # Filter only localhost script requests
    script_df = filter_localhost_scripts(df)

    # Display top script data
    print(script_df[["url", "transferSizeKB"]])  # Print table in console

    # Generate plot
    if not script_df.empty:
        plot_top_localhost_scripts(script_df)
    else:
        print("No matching script requests found from localhost:3000 or localhost:8080.")

if __name__ == "__main__":
    main()
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate  # Import tabulate for nicely formatted tables

def load_lighthouse_json(file_path):
    """Loads Lighthouse JSON and extracts network request data."""
    with open(file_path, "r", encoding="utf-8") as file:
        data = json.load(file)

    # Extract network requests
    network_requests = data["audits"]["network-requests"]["details"]["items"]
    
    # Convert data to DataFrame
    df = pd.DataFrame(network_requests)

    # Convert transferSize from bytes to KB
    df["transferSizeKB"] = df["transferSize"] / 1024

    return df

def filter_localhost_requests(df):
    """Filters network requests for localhost:3000 and localhost:8080, sorting by size."""
    filtered_df = df[
        df["url"].str.startswith(("http://localhost:3000", "http://localhost:8080"))
    ].copy()

    # Sort by transfer size descending
    filtered_df = filtered_df.sort_values(by="transferSizeKB", ascending=False)

    return filtered_df

def plot_transfer_size_by_resource(df):
    """Bar chart of total transfer size by resource type for localhost requests."""
    plt.figure(figsize=(10, 5))
    df.groupby("resourceType")["transferSizeKB"].sum().plot(kind="bar")
    plt.xlabel("Resource Type")
    plt.ylabel("Total Transfer Size (KB)")
    plt.title("Total Transfer Size by Resource Type (localhost:3000 & 8080)")
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle="--", alpha=0.7)
    plt.show()

def plot_resource_distribution(df):
    """Pie chart of transfer size distribution by resource type for localhost requests."""
    plt.figure(figsize=(8, 8))
    df.groupby("resourceType")["transferSizeKB"].sum().plot(kind="pie", autopct='%1.1f%%')
    plt.ylabel("")
    plt.title("Distribution of Transfer Size by Resource Type (localhost:3000 & 8080)")
    plt.show()

def plot_top_localhost_requests(df):
    """Bar chart for top individual requests from localhost by transfer size."""
    plt.figure(figsize=(12, 6))
    plt.barh(df["url"], df["transferSizeKB"], color="blue")
    plt.xlabel("Transfer Size (KB)")
    plt.ylabel("URL")
    plt.title("Top Individual Requests from localhost:3000 & 8080 by Data Usage")
    plt.gca().invert_yaxis()  # Invert y-axis to have the largest on top
    plt.grid(axis="x", linestyle="--", alpha=0.7)
    plt.show()

def print_table(df):
    """Prints the filtered localhost requests as a copy-friendly table."""
    table_data = df[["url", "resourceType", "transferSizeKB"]]
    print(tabulate(table_data, headers="keys", tablefmt="grid"))

def main():
    # Provide the path to your Lighthouse JSON report
    file_path = "lighthouse_report_page.json"

    # Load data
    df = load_lighthouse_json(file_path)

    # Filter only localhost requests
    localhost_df = filter_localhost_requests(df)

    # Display top localhost data usage in a copy-friendly table
    if not localhost_df.empty:
        print_table(localhost_df)
        plot_transfer_size_by_resource(localhost_df)
        plot_resource_distribution(localhost_df)
        plot_top_localhost_requests(localhost_df)
    else:
        print("No matching requests found for localhost:3000 or localhost:8080.")

if __name__ == "__main__":
    main()

```
